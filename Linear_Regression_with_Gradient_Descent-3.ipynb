```python
import numpy as np
import matplotlib.pyplot as plt

# Generate example data
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Linear regression using gradient descent
def gradient_descent(X, y, theta, learning_rate, num_iterations):
    m = len(X)
    cost_history = []
    theta_history = []

    for iteration in range(num_iterations):
        gradients = (1/m) * X.T.dot(X.dot(theta) - y)
        theta = theta - learning_rate * gradients
        cost = (1/(2*m)) * np.sum((X.dot(theta) - y)**2)
        cost_history.append(cost)
        theta_history.append(theta.copy())

    return theta, cost_history, theta_history

# Add a bias term (intercept) to the features
X_b = np.c_[np.ones((100, 1)), X]

# Initialize coefficients (intercept and slope)
theta = np.random.rand(2, 1)

# Define hyperparameters
learning_rate = 0.1
num_iterations = 1000

# Perform gradient descent
optimal_theta, cost_history, theta_history = gradient_descent(X_b, y, theta, learning_rate, num_iterations)

# Make predictions
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(optimal_theta)

# Plot the data and regression line
plt.scatter(X, y, label='Data Points')
for i in range(len(theta_history)):
    y_predict_iter = X_new_b.dot(theta_history[i])
    plt.plot(X_new, y_predict_iter, 'r-', alpha=0.2)
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()


plt.scatter(X, y, label='Data Points')
plt.plot(X_new, y_predict, 'r-', label='Regression Line', linewidth=2)
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

```
